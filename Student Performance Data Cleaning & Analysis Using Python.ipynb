```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.loc[:, "name"] = df["name"].str.lower().str.title()
```

```python
df
```

```python
df["age"].unique()
```

```python
age_mean = df["age"].mean().round(0)
```

```python
df.loc[:, "age"] = df["age"].fillna(age_mean)
```

```python
df["gender"].unique()
```

```python
df.loc[:, "gender"] = df["gender"].str.lower().str.title()
```

```python
df
```

```python
df["grade"].unique()
```

```python
df.loc[:, "grade"] = df["grade"].replace({'10th':10, '11th':11})
```

```python
df
```

```python
math_mean = df["math_score"].mean()
```

```python
df.loc[:, "math_score"] = df["math_score"].fillna(math_mean)
```

```python
df
```

```python
df["english_score"]
```

```python
df.loc[:, "english_score"] = pd.to_numeric(df["english_score"], errors = 'coerce')
```

```python
english_mean = df["english_score"].mean()
```

```python
df.loc[:, "english_score"] = df["english_score"].fillna(english_mean)
```

```python
df["english_score"] = pd.to_numeric(df["english_score"], errors='coerce')
df["english_score"] = df["english_score"].astype(float).round(0)
```

```python
df
```

```python
df["enrolled_date"].unique()
```

```python
dict_date = {
    '10-06-2022' : '2022-06-10', 
    '06/12/2022' : '2022-06-12',
    '2022/06/11' : '2022-06-11'
}
```

```python
df.loc[:, "enrolled_date"] = df["enrolled_date"].replace(dict_date)
```

```python
df
```

```python
df.loc[:, "enrolled_date"] = pd.to_datetime(df["enrolled_date"], dayfirst = True)
```

```python
df
```
